```python
#Define function to check outlier in a variable
def data_outlier(x):
    q1 = x.quantile(0.25)
    q2 = x.quantile(0.50)
    q3 = x.quantile(0.75)
    iqr = q3-q1
    ul = q3+1.5*iqr
    ll = q1-1.5*iqr
    below = (x<=ll).sum()
    above = (x>=ul).sum()
    return below, above

#Calculate outliers in all variables in a dataset
def count_outlier(x):
    res = {}
    for i in x.select_dtypes(include="number").columns: #only select the numeric variables
        res[i] = data_outlier(x[i])
    return res
```

```python
#Check outliers in the dataframe
total_out = {}
for name,data in df.items():
    total_out[name] = count_outlier(data)
print(total_out)
```

```python
#Q1:Average listing price by city
query = """
select City,avg(price) as Average_Price
from listing
group by city
order by Average_Price
"""
pd.read_sql(query,con=engine)
```

```python
#Q2:Average price per square foot by property type
query = """
select Property_Type, avg(Price/Sqft) as Average_Price_persqft
from listing
group by Property_type
order by Average_Price_persqft
"""
pd.read_sql(query,con=engine)
```

```python
#Q3: How Furnishing status impact property prices
query = """
select furnishing_status, avg(Price) as Price
from listing
inner join prop_attri
on listing.Listing_ID=prop_attri.listing_id
group by furnishing_status
order by Price desc;
"""
pd.read_sql(query,con=engine)
```

```python
#Q4: Do property closer to metro stations command higher prices
query = """ 
select metro_distance_km as Metro_Distance,avg(price) as Average_Price
from listing
inner join prop_attri
on listing.Listing_ID=prop_attri.listing_id
group by metro_distance_km
order by metro_distance_km, Average_Price desc;
"""
pd.read_sql(query,con=engine)
```

```python
#Q5: Rented properties price differently from non-rented ones
query = """ 
select is_rented as Rented, avg(Price) as Average_Price
from listing
inner join prop_attri
on listing.Listing_ID=prop_attri.listing_id
group by Rented
order by Average_Price desc;
"""
pd.read_sql(query,con=engine)
```

```python
#Q6: How do bed rooms and bath rooms affect price
query = """ 
select bedrooms as Bedroom, bathrooms as Bathroom, avg(Price) as Average_Price
from listing
inner join prop_attri
on listing.Listing_ID=prop_attri.listing_id
group by Bedroom, Bathroom
order by Bedroom, Average_Price asc
"""
pd.read_sql(query,con=engine)
```

```python
#Q7: Properties with parking and powerbackup selling at higher costs
query = """ 
select parking_available as Parking, power_backup as Power_backup, avg(price) as Avg_Price
from listing
inner join prop_attri
on listing.Listing_ID=prop_attri.listing_id
group by Parking, Power_backup
order by Parking, Avg_Price asc
"""
pd.read_sql(query,con=engine)
```

```python
#Q8: Year built influence listing price
query = """ 
select year_built as Year, avg(Price) as Avg_Price
from listing
inner join prop_attri
on listing.Listing_ID=prop_attri.listing_id
group by Year
order by Year asc
"""
pd.read_sql(query,con=engine)
```

```python
#Q9: Which Cities has the highest median property prices
query = """ 
with med as (
select 
    City, 
    Price,
    row_number() over(partition by City order by Price) as Rownum,
    count(*) over(partition by City) as total_count
from listing
)
select City, avg(Price) as Median_Value
from med 
where Rownum in (floor((total_count+1)/2),ceil((total_count+1)/2))
group by City
order by avg(Price) asc
"""
pd.read_sql(query,con=engine)
```

```python
#Q10: How are the properties distributed across price buckets
query = """ 
select quartile, avg(Price) as Avg_Price
from (select Price,
Ntile(4) over(order by Price) as quartile
from listing
) as sub
group by quartile
order by quartile
"""
pd.read_sql(query,con=engine)
```

```python
##Sales and market performance
#Q11: Average days on market by city
query = """ 
select City, avg(Days_on_Market) as Avg_days
from listing
inner join sales
on listing.Listing_ID=sales.Listing_ID
group by City
order by Avg_days
"""
pd.read_sql(query,con=engine)
```

```python
#Q12 Which Property types sell the fast?
query = """ 
with pr_fa as (select s.Date_Sold, l.Date_Listed, l.Property_Type
from sales s
inner join listing l
on l.Listing_ID=s.Listing_ID
)
select Property_Type, Date_Listed, Date_Sold, 
    datediff(Date_Sold, Date_Listed) as days
from pr_fa
order by days desc;
"""
pd.read_sql(query,con=engine)
```

```python
#Q13 What percentage of properties are sold above listing price
query = """ 
with pr_fa as (select s.Sale_Price, l.Price
from sales s
inner join listing l
on l.Listing_ID=s.Listing_ID
)
select
(sum(case when Sale_Price > Price then 1 else 0 end)*100/count(*)) as per
from pr_fa
"""
pd.read_sql(query,con=engine)
```

```python
#Q14 What is the sale to list price ratio by city?
query = """
with pr_fa as (select s.Sale_Price, l.Price, l.City
from sales s
inner join listing l
on l.Listing_ID=s.Listing_ID
)
select City, avg(Sale_Price/Price) as Ratio
from pr_fa
group by City
order by Ratio
"""
pd.read_sql(query, con=engine)
```

```python
#Q15 Which listing took more than 90 days to sell
query = """ 
with pr_fa as (select s.Date_Sold, l.Date_Listed, l.City, l.Listing_ID
from sales s
inner join listing l
on l.Listing_ID=s.Listing_ID
)
select Listing_ID, City,
    datediff(Date_Sold, Date_Listed) as days
from pr_fa
where datediff(Date_Sold, Date_Listed)>90;
"""
pd.read_sql(query, con=engine)
```

```python
#Q16 How does metro distance affect time on market?
query = """ 
with m_d as (select p.listing_id, p.metro_distance_km, s.Sale_Price
from prop_attri p 
inner join sales s
on p.listing_id=s.Listing_ID)
select listing_id, metro_distance_km, Sale_Price
from m_d
order by metro_distance_km, Sale_Price
"""
pd.read_sql(query, con=engine)
```

```python
#Q17 What is the monthly sales trend?
query = """ 
select DATE_FORMAT(Date_Sold, '%%m') as month, count(*) as Total_Sales
from sales
group by DATE_FORMAT(Date_Sold, '%%m')
order by month
"""
pd.read_sql(query,con=engine)
```

```python
#Q18 Which properties are currently unsold?
query = """ 
with pr_un as (select l.Listing_ID, l.Property_Type, l.Date_Listed, s.Date_Sold
from Listing l
left join sales s
on l.Listing_ID=s.Listing_ID)
select Property_Type, count(*) as Total_unsold
from pr_un
where Date_Sold is NULL
group by Property_Type
order by Total_unsold
"""
pd.read_sql(query, con=engine)
```

```python
#Q19: Which agents have closed the most sales
query = """  
select ag.Agent_ID, ag.Name, ag.Email as Email, ag.deals_closed as Deal
from agents ag
order by Deal desc
limit 5
"""
pd.read_sql(query, con=engine)
```

```python
#Q20: Top agents by total sales revenue
query = """ 
with revenue as (select ag.Agent_ID, ag.Name, ag.Email, l.Listing_ID
from agents as ag
inner join listing as l
on l.Agent_ID = ag.Agent_ID)
select r.Agent_ID, r.Name, r.Email, s.Sale_Price
from revenue as r
inner join sales as s
on s.Listing_ID=r.Listing_ID
order by s.Sale_Price desc
limit 10;
"""
pd.read_sql(query,con=engine)
```

```python
#Q21: Which agents close deals fast
query = """ 
select ag.Agent_ID, ag.Name, ag.Email, ag.avg_closing_days
from agents ag
order by ag.avg_closing_days asc
limit 5
"""
pd.read_sql(query,con=engine)
```

```python
#Q22: Does experience correlated with deals closed
query = """ 
select experience_years as Exp, avg(deals_closed) as Avg_Deals_closed
from agents
group by Exp
order by Exp asc
"""
pd.read_sql(query,con=engine)
```

```python
#Q23 Do agents with high rating close deal faster
query = """ 
select rating, avg(deals_closed) as Avg_Deals_closed
from agents
group by rating
order by rating
"""
pd.read_sql(query,con=engine)
```

```python
#Q24: What is the average commission earned by each agent
query = """ 
select ag.Agent_ID, ag.Name, avg(ag.commission_rate) as Commission_rate
from agents as ag
group by ag.Agent_ID,ag.Name
order by Commission_rate
"""
pd.read_sql(query,con=engine)
```

```python
#Q25: Which agents have the most active listing
query = """ 
select Agent_ID, count(*) as total_listing
from listing
group by Agent_ID
order by total_listing
limit 1
"""
pd.read_sql(query, con=engine)
```

```python
#Q26: What percentage of buyers are investors vs end users?
query = """ 
select
(sum(case when buyer_type = 'Investor' then 1 else 0 end)*100/count(*)) as percentage
from buyers
"""
pd.read_sql(query,con=engine)
```

```python
#Q27: Which cities have the highest loan uptake
query = """
with c_l as (select b.loan_taken, s.Listing_ID
from buyers as b
inner join sales as s
on s.sale_id=b.sale_id)
select l.City, count(b.loan_taken) as Loan_taken
from c_l as b
inner join listing as l
on l.sale_id=b.sale_id
group by City
order by Loan_taken desc
limit 10;
"""
#Note there is no sale_id in the sales dataset
```

```python
#Q28: Average loan amount by buyer type
query = """ 
select buyer_type as Buyer_type,avg(loan_amount) as Avg_Loan_amt
from buyers
group by buyer_type
order by Buyer_type
"""
pd.read_sql(query,con=engine)
```

```python
#Q29: Which payment mode is commonly used
query = """ 
select payment_mode as Payment_mode,count(*) as Frequency
from buyers
group by Payment_mode
order by Frequency
"""
pd.read_sql(query,con=engine)
```
